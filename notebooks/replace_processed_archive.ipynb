# processed archive 교체 노트북 (1회용)

기존 Drive의 `processed_data.tar.gz`를 삭제하고, 현재 세션의 `/content/data/processed`를 다시 압축해 같은 경로에 저장합니다.

- 이 노트북은 **버그 수정 이전 archive를 교체할 때만** 사용하세요.
- `download_to_drive.ipynb`의 Step 0~2가 이미 끝나서 `/content/data/processed`가 존재하는 상태를 가정합니다.
- 실수 방지를 위해 아래 코드에서 `CONFIRM_REPLACE = True`로 직접 바꿔야만 삭제/저장이 진행됩니다.


```python
import os

DRIVE_ROOT    = '/content/drive/MyDrive/Colab Notebooks/ARMY Projects/2026-tactical-speech-enhancement'
DRIVE_DATA    = f'{DRIVE_ROOT}/data'
DRIVE_ARCHIVE = f'{DRIVE_DATA}/processed_data.tar.gz'
LOCAL_ARCHIVE = '/tmp/processed_data.tar.gz'
LOCAL_DATA    = '/content/data/processed'

print(f'Drive archive: {DRIVE_ARCHIVE}')
print(f'Local data:     {LOCAL_DATA}')
print(f'Local tar:      {LOCAL_ARCHIVE}')
```

```python
CONFIRM_REPLACE = False  # 실제 교체 시 True로 바꾸세요

if not CONFIRM_REPLACE:
    raise RuntimeError('CONFIRM_REPLACE = True 로 바꾼 뒤 다시 실행하세요.')

if not os.path.isdir(LOCAL_DATA):
    raise FileNotFoundError(f'로컬 processed 데이터가 없습니다: {LOCAL_DATA}')

os.makedirs(DRIVE_DATA, exist_ok=True)
print('확인 완료: 교체를 시작합니다.')
```

```python
if os.path.exists(DRIVE_ARCHIVE):
    print(f'기존 archive 삭제: {DRIVE_ARCHIVE}')
    !rm -f "{DRIVE_ARCHIVE}"
else:
    print('기존 archive 없음 — 새로 저장합니다.')

print('① 로컬에서 새 archive 압축 중...')
!tar -czf "{LOCAL_ARCHIVE}" -C /content/data processed
!du -sh "{LOCAL_ARCHIVE}"

print('② Drive로 복사 중...')
!cp "{LOCAL_ARCHIVE}" "{DRIVE_ARCHIVE}"
!rm "{LOCAL_ARCHIVE}"

!ls -lh "{DRIVE_DATA}"
print('\n✅ archive 교체 완료')
```
